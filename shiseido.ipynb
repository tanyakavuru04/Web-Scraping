```python
import requests
import re
import pandas as pd
```

```python
from bs4 import BeautifulSoup
```

```python
baseurl = 'https://www.shiseido.com/us/en/sunscreen/'
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
```

```python
r = requests.get(baseurl)
```

```python
soup = BeautifulSoup(r.content, 'lxml')
```

```python
productlist = soup.find_all('div', class_ = 'small-12 medium-4 large-3 column gtmindex')
```

```python
print(productlist)
```

```python
productlinks = []
```

```python
# for link in item.find_all('a', href=True):
#     productlinks.append(baseurl + link['href'])
```

```python
for item in productlist:
  for link in item.find_all('a', href = True):
    productlinks.append(baseurl + link['href'])
```

```python
#testlink = 'https://www.shiseido.com/us/en/urban-environment-vita-clear-sunscreen-spf-42-0730852195479.html?cgid=suncare'
all_sunscreens = []
# for link in productlinks:
for link in list(set(productlinks)):
  r = requests.get(link, headers=headers)
  soup = BeautifulSoup(r.content, 'lxml')
  name = soup.find('h1', class_ = 'product-name shi-pri-text-color-4').text.strip()
  price = soup.find('span', class_ = 'price-sales-range hide').text.strip()
  description = soup.find(class_ = 'product-description shi-pri-text-color-4').text
  product = soup.find('div', class_ = 'product-title column medium-6 large-5')
  try:
    rating = product.find('span', class_ = 'bvseo-ratingValue').text.strip()
  except:
    rating = 'Rating not available'
  try:
    number_reviews = product.find('span', class_ = 'bvseo-reviewCount').text.strip()
  except:
    number_reviews = 'Number of reviews not available'
  keying = soup.find('div', class_ = 'pdp-content-inner row shi-row')
  try:
    howtouse = keying.find_all('div', class_ ='row')[0].text.strip()
  except:
    howtouse = 'exception error'
  try:
    size = soup.find(class_ = 'swatches-option selectable selected').text.strip()
  except:
    size = 'Size not available'
  try:
    results = keying.find_all('ul')[0].text.strip()
  except:
    results = 'Results not available'
  try:
    benefits = keying.find_all('ul')[1].text.strip()
  except:
    benefits = 'Benefits not available'
  try:
    ingredients = keying.find_all('div', class_ ='row')[1].text.strip()
  except:
    ingredients = 'Ingredients not available'
  sunscreens = {'Name':name,
                'Price':price,
                'Size':size,
                'Rating': rating,
                'Number of reviews': number_reviews,
                'Description':description,
                'Results':results,
                'Benefits':benefits,
                'How to Use':howtouse,
                'Key Ingredients':ingredients
                }
  all_sunscreens.append(sunscreens)
```
